### 第一题
已知实验中学举行了高二期中模拟考试，考试后分别计算了全体男生、女生各科的平均成绩，统计结果如表所示：

https://cdn2.kaoshiyun.com.cn/File/360593/Question/d7280b7/image/20220914165843_2477.png

现在班主任老A想要直观对比不同性别的学生平均成绩的差异，请选择合适的统计图进行可视化。并为图形添加组件：标题、x轴标签、y轴标签、图例、文本注释等

```python
# 1.准备数据
# x
course = ["语文", "数学", "英语", "物理", "化学", "生物"]
# y_1
boy_score = [85.5, 91, 72, 59, 66, 55]
# y_2
girl_score = [94, 82, 89.5, 62, 49, 53]

# 2.可视化 ->> 成绩差异 ->> 条形图
plt.figure(figsize=(14,8))

# 2.1设置原始柱子的位置,0 1 2
p = np.arange(len(course))

# 2.2设置柱子的宽度 ->> 2*width<1
width = 0.3
plt.bar(p-width/2,boy_score,width=width,label="男生")
plt.bar(p+width/2,girl_score,width=width,label="女生")

# 3.其他组件
plt.xticks(p,course)
plt.xlabel("学科")
plt.ylabel("成绩")
plt.title("高二男、女生平均成绩一览表")
plt.legend()

# 4.添加注释
# 方法1：plt.annotate()
# for x,y_1,y_2 in zip(p,boy_score,girl_score):
#     plt.annotate(f"{y_1}",(x-width/2,y_1),((x-width/2)-0.06,y_1))
#     plt.annotate(f"{y_2}",(x+width/2,y_2),((x+width/2)-0.06,y_2))

# 方法2：plt.text()   添加文本
# x,y:文本所在位置
# s:文本
for x,y_1,y_2 in zip(p,boy_score,girl_score):
    plt.text(x-width/2,y_1,y_1,horizontalalignment='center')
    plt.text(x+width/2,y_2,y_2,horizontalalignment='center')
    
plt.show()
```

### 第二题

按照如下要求绘制图表：

（1） 画布被规划为2*3的矩阵区域；

（2） 在编号为3的区域中绘制包含一条正弦曲线的子图；

（3） 在编号为6的区域中绘制包含一条余弦曲线的子图；

（4） 共享两个子图的x轴。

```python
# 1.构建x,y = sin(x)
x = np.linspace(-np.pi,np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

# 2.可视化
fig,axes = plt.subplots(2,3,sharex = True)    # sharex 共享x轴
axes[0][2].plot(x,y1)
axes[1][2].plot(x,y2)

plt.show()
```
