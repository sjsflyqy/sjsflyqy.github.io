```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 步骤一（替换sans-serif字体）
plt.rcParams['axes.unicode_minus'] = False
```

### 直方图

```python
# x:绘图的数据
# bins:组数;注意：一定整数
# density:True ->> 频率图

# 1.构建数据
temp_li= [6.9,4.1,6.6,5.2,6.4,7.9,8.6,3.0,4.4,6.7,7.1,4.7,9.1,6.8,8.6,5.2,5.8,7.9,5.6,8.8,8.1,5.7,8.4,4.1,6.4,6.2,5.2,6.8,5.6,5.6,6.8,8.2,6.4,4.8,6.9,7.1,9.7,6.4,7.3,6.8,7.1,4.8,5.8,6.5,5.9,7.3,5.5,7.4,6.2,7.7]

# 2.计算组数
# 2.1 计算最大值与最小值的差
cha = max(temp_li) - min(temp_li)
# 2.2初始化组距
b = 1
# 2.3组数
bins_ = int(round(cha)/b)
bins_

# 3.可视化
plt.hist(temp_li,bins=bins_,density=True)

# 5.设置组件
plt.xlabel("温度分布区间")
plt.ylabel("频数")
plt.title("温度分布直方图")

# 4.展示图形
plt.show()
```

### 扇形图
将以下frac数据绘制扇形图，并且设置其扇形标签为label

frac = [1/50,6/50,11/50,15/50,9/50,6/50,2/50]

label = ['[3,4]','(4,5]','(5,6]','(6,7]','(7,8]','(8,9]','(9,10]']

```python
# 参数
# x:数据情况,它的占比，和为1
# explode:设置某个分块是否要分离饼图
# labels:指定标签
# autopct:百分比数据标签
# shadow:是否有阴影部分
# plt.pie

# 返回值
# patches:每一个扇形对象
# texts:label的文本列表
# autotexts:百分比的文本列表

# 1.构建数据
frac = [1/50,6/50,11/50,15/50,9/50,6/50,2/50]
label = ['[3,4]','(4,5]','(5,6]','(6,7]','(7,8]','(8,9]','(9,10]']
explode_ = [0,0,0,0.1,0,0,0]

# 2.可视化 
# plt.pie(frac,labels=label,autopct="%1.2f%%",explode=explode_,shadow=True)
# 不加返回值也可以生成图形
patches,texts,autotexts = plt.pie(frac,labels=label,autopct="%1.2f%%",explode=explode_,shadow=True)

# 3.需求：设置百分比字体的颜色显示为绿色
for autotext in autotexts:
    autotext.set_color("g")
    
# 4.展示图形
plt.show()
```

### 雷达图

```python
# 安装numpy,pip install numpy
# 导入numpy
import numpy as np

# 第一个参数：theta->>x轴坐标点为弧度
# 注意：360°->>角度  2*PI->>弧度 
# 第二个参数：数据

# 1.准备数据
quaters = ['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q1']  # 注意：这不是x，x是弧度  
sales = [40,91,44,90,20,54,80]

# 2.求弧度。设置每个扇区弧度，将360°等分
#求的数据个数
n = len(sales)
# 创建等差数列
angels = list(np.linspace(0, 2*np.pi,n))

# 想要图形是闭合得区间，起始值=终点值
# 数据再处理，使得最后一个数据与第一个数据连接起来
angels.append(angels[0])
sales.append(sales[0])

plt.polar(angels,sales,">-",color="g")
# plt.fill(angels,sales,facecolor="g")  填充雷达图内部颜色

# 设置刻度标签
plt.xticks(angels,quaters)

plt.show()
```

### 箱型图

```python
import random

# 随机生成100个1-100得数据
data = [random.randint(0,100) for i in range(100)]
data

# 添加2个异常数据
data.extend([-200,500])
data

# 绘制箱型图
# x:数据
# notch:置信区间  用的少，基本不用
# sym:异常点得符号
# meanline:设置均线
# showmeans:显示均值
# vert:False设置为水平箱型图

plt.boxplot(data,notch=True,sym="^",meanline=True,showmeans=True,vert=False)
plt.show()
```
