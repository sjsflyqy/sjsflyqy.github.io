```python
import matplotlib.pyplot as plt
```

```python
# 1.创建画布
fig = plt.figure(figsize=(14,8))
fig


# 2.添加子图对象到当前画布中
# - nrows:行
# - ncols:列
# - index:子图的索引
ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)


# 3.绘画，基于axes对象
ax1.scatter([1,2,3,5],[2,4,5,6])
# ax2.scatter([1,2,3,5],[2,4,5,6])
ax3.scatter([1,2,3,5],[2,4,5,6])
ax4.scatter([1,2,3,5],[2,4,5,6])
plt.show()
```

```python
# plt.subplots   ->> 创建画布对象并且设置子图，返回值：画布对象、子图对象
# nrows:子图的行数
# ncols:子图的列数
fig,axes = plt.subplots(2,2)
axes[0][0].scatter([1,2,3,4],[5,6,7,8])
axes[1][1].scatter([1,2,3,4],[5,6,7,8])
plt.show()
```

```python
# 注意：不用子图时，直接使用figure即可,甚至直接一个scatter就可以了
plt.figure()
plt.scatter([1],[3])
plt.show()
```
