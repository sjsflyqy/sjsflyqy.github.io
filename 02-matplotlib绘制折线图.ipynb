```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [5,6,10,8]
plt.plot(x,y,"r",marker="o",linestyle="--")
plt.show()
```

```python
fig,axes = plt.subplots(2,2)
axes[0][0].plot([1,2,4,6],[4,3,6,8],'r',marker='.',linestyle="--")
axes[0][1].plot([1,2,4,6],[4,3,6,8],'g',marker='o',linestyle="--")
axes[1][0].plot([1,2,4,6],[4,3,6,8],marker='x',linestyle="--")
plt.show()
```
